# Contrast Adjustment pada Pengolahan Citra
Notebook ini melakukan penyesuaian kontras (Contrast Adjustment) menggunakan OpenCV.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Baca gambar
img = cv2.imread('img-3.jpg')
img
```

```python
if img is None:
    print(" Gambar tidak ditemukan! Pastikan file ada di folder yang sama.")
else:
    # Konversi BGR  RGB
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # ====== Contrast Adjustment ======
    # alpha = nilai kontras (>1 untuk lebih kontras, 0-1 untuk kurang kontras)
    alpha = 1.5  # ubah nilai ini untuk melihat efek berbeda
    contrast_img = cv2.convertScaleAbs(img_rgb, alpha=alpha, beta=0)

    # ====== Tampilkan Hasil ======
    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    plt.title("Original")
    plt.imshow(img_rgb)
    plt.axis("off")

    plt.subplot(1, 2, 2)
    plt.title("Contrast Adjusted")
    plt.imshow(contrast_img)
    plt.axis("off")

    plt.show()
```
